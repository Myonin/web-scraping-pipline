```python
def _clean_text(page):

    page = re.sub('\n', '', page)
    page = re.sub(' +', ' ', page)
    page = re.sub('Пристрелить \(\+\)', 'END', page)

    return page

def _get_content(_regex, page):
    page_content = {}
    for key in _regex.keys():
        page_content[key] = re.findall(_regex[key], page)

    return page_content
```
